```python
## Dependencies
import csv
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import requests
import json
```

```python
## Analysis
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "Imperial"
weather_api_key = "6d0518d886aac862895d41dd1afecfae"
```

```python
## Generate Cities List

from citipy import citipy
from random import *
cities = []

while(True):
    city = citipy.nearest_city(uniform(-90, 90), uniform(-180, 180))
    if city not in cities:
        cities.append(city.city_name)
    if len(cities) == 500:
        break
```

```python
## Perform API Calls
query_url = f"{url}units={units}&appid={weather_api_key}&q="
for city in cities:
    response = requests.get(query_url + city).json()
    #print(query_url + city)
    
```

```python
## City Weather Data Frame

# set up lists to hold reponse info
xcity = []
cloudiness = []
country = [] 
date = []
humidity = []
lat = [] 
lng = [] 
max_temp = [] 
wind_speed = []

for city in cities:
    response = requests.get(query_url + city).json()
    xcity.append(response['name'])
    cloudiness.append(response['clouds']['all'])
    country.append(response['sys']['country'])
    date.append(response['dt'])
    humidity.append(response['main']['humidity'])
    lat.append(response['coord']['lat'])
    lng.append(response['coord']['lon'])
    max_temp.append(response['main']['temp_max'])
    wind_speed.append(response['wind']['speed'])
    
#Create a Data Dictionary 
weather_dict = {
    "City": xcity,
    "Cloudiness": cloudiness,
    "Country": country,
    "Date": date,
    "Humidity": humidity,
    "Lat": lat,
    "Lng": lng,
    "Max Temp": max_temp,
    "Wind Speed": wind_speed }

#Create a Data Frame 
weather_data = pd.DataFrame(weather_dict)

#Display how many values in each column in Weather Data Frame
weather_data.count()
```

```python
# display first 5 rows in Weather Data Frame
weather_data.head()
```

```python
## Latitude vs Temperature Plot

plt.scatter(x= weather_data["Lat"], y= weather_data["Max Temp"], c="blue", edgecolor='black', lw=1, alpha=0.8)

# adding the required limit, style, title, legend, and text
plt.xlim(-80, 100)
plt.ylim(-20,120)
plt.title("City Latittude vs. Temperature  (01/05/2017)" , fontsize= 14)
plt.xlabel("Latittude ")
plt.ylabel("Max Temperature (F)")
mpl.style.use("seaborn")
plt.savefig("Latitude vs Temperature Chart.png")

# display the final plot

plt.show()
```
